# 1. The Smart Attendance Logger
## The Goal: Manage a classroom attendance list and ensure no duplicate entries are recorded if a student taps their ID card twice.
        1. Concepts: set , function , while .
        2. The Problem: Create a function mark_attendance() that keeps asking for student names until the user
        types "STOP". Store names in a Set to automatically prevent duplicates.
        3. Input: Multiple strings (names) entered one by one.
        4. Output: The final count of unique students and the set itself.
        5. Hint:
        Expected Flow: Start an empty set Use a while True loop to take input() if name is "STOP", break
        else add name to set Finally, print len(attendance_set) 

```python
def mark_attendens():
    student_set = set()
    Studen_name =""
    while  True:
        Studen_name= input("Scan your ID")
        Studen_name = Studen_name.strip().upper()

        if Studen_name == 'STOP':
            break
            
        student_set.add(Studen_name)
        print('ID sucessfully scaned')
    print(student_set)
    print("Number of students present are : ",len(student_set))
        
```

```python
mark_attendens()
```


# 2 The Restaurant Bill Splitter
## The Goal: Calculate how much each person owes, including a tax and a custom tip percentage.
        Concepts: functions , default values , list .
        The Problem: Write a function split_bill(items_prices, people, tax=0.05, tip=0.10) .
        items_prices will be a List of decimals (floats).
            1. Input: A list like [25.0, 15.0, 10.0] and an integer for people .
            2. Output: A single float (the amount per person).
        Hint: Expected Flow: Calculate sum() of the list Add tax and tip amounts Divide the grand total by people return the result.

```python
def split_calculation():
    price_of_items = []
    base_bill = 0
    tax = 0.05
    while True:
        waiter_input = float(input('Enter amount of each item OR 0 to exit'))
        if waiter_input ==0:
            break
        
        price_of_items.append(waiter_input)
    print(price_of_items)    
    base_bill = sum(price_of_items)
    print("Bill before tax:", base_bill)

    total_bill = base_bill + (base_bill * tax)
    print("Bill after tax of 5% :", total_bill)

    numer_of_ppl = float(input("Enter number of people to split bill"))
    amount_per_person = total_bill / numer_of_ppl

    print("Amount per preson is :",amount_per_person)
```

        2. Ask the user for 3 items they want to wear. Check each item individually against the list.
        
        Input: Three separate input() strings (e.g., "Red Tie", "White Shirt", "Socks").
        Output: A "Missing" warning for every item not found in the list.
        Hint: Expected Flow: Take 3 inputs Use three separate if statements (not elif ) to check if each input is not in
        the wardrobe list Print "Missing [Item]" for each failure.

```python
def digital_wardrobe():
    wardrobe = ["Blue Jeans", "White Shirt", "Black Jacket"]
    check_list = []
    missing_item = []
    

    for i in range(len(wardrobe)):
        wardrobe[i]=wardrobe[i].strip().lower()
        check_list.append(wardrobe[i])


    user_input1 = input("Enter your first chosse of cloth")
    user_input1 = user_input1.strip().lower()
    user_input2 = input("Enter your second chosse of cloth")
    user_input2 = user_input2.strip().lower()
    user_input3 = input("Enter your third chosse of cloth")
    user_input3 = user_input3.strip().lower()

    if user_input1 not in check_list:
        missing_item.append(user_input1)
        print("Item missing : ",user_input1)

    if user_input2 not in check_list:
        missing_item.append(user_input2)
        print("Itme missing : ",user_input2)

    if user_input3 not in check_list:
        missing_item.append(user_input3)
        print("Itme missing : ",user_input3)
    return 'Missing items aree ',missing_item
```

# 7. The ATM Pin Validator
## The Goal: Lock the user out after 3 incorrect attempts.
    Concepts: while , if-else , break .
    The Problem: Set correct_pin = "1234" . Allow the user 3 tries to enter the right PIN.
    Input: Up to 3 string inputs.
    Output: "Welcome" on success, or "Card Blocked" after the 3rd fail.
    Hint: Expected Flow: Set attempts = 0 While attempts < 3 : Take PIN input if correct, print "Welcome" and break else increment attempts . If attempts reaches 3, print "Card Blocked.

```python
def num_of_pin_entry_limit():
    valid_pin = 1234
    
    set_attempts = 0
    while set_attempts <3:
        user_input_pin = int(input("Enter valid pin"))
        if user_input_pin ==valid_pin:
            print("Pin entered successfullly")
            break
        else:
            set_attempts +=1
            #call block card pin
            print("Please enter correct pin")
            
    if set_attempts == 3:
        print("Card has been blocked")
```

# 9. The Grocery Inventory Update
## The Goal: Update the quantity of items in a store after a shipment arrives.
    Concepts: dict ,  for loop.
    The Problem: Given_inventory = {"Apples": 50, "Bananas": 20} , 
    A shipment of 10 more of each arrives. Update the dictionary using a loop.
    Input: None.
    Output: The updated dictionary.
    Hint:
    Expected Flow: Loop through inventory.keys() For each item, add 10 to the current value Print the dictionary after the loop finishes.

```python
def grocery_inventory_update():
    Given_inventory = {"Apples": 50, "Bananas": 20} 
    updated_inventory = {}
    for i, j in Given_inventory.items():
        updated_inventory[i]= j+10
    grocery_inventory_update = updated_inventory
    print("10 Items has been added to inventory")
    print(grocery_inventory_update)

```

# 10. The Flexible Event Invitation
## The Goal: Create a function that can accept any number of guests and specific event details.
    Concepts: functions , *args , **kwargs .
    The Problem: Write create_invite(event_name, *guest_names, **details) .
    Input: One string, followed by multiple guest names, followed by named details (e.g., 
    Output: A printed invitation showing the event, guests, and details.
    Hint:Expected Flow: Print the  event_name Use a loop through **details (extra info) and print them.

```python
def create_invite(event_name, *guest_names, **details):
    print(f"You're Invited to: {event_name}")
    print("\nGuest List:")

    for guest in guest_names:
        print("-", guest)

    print("\nEvent Details:")
    for key, value in details.items():
        print(f"{key.capitalize()}: {value}")
```
