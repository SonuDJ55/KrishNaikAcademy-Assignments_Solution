1. The Office Door Badge You are programming a smart door for a tech office. The door should only open if the
user provides the correct "Access Code."
Input: User enters a 4-digit code.
Scenario: If the code is 
1234 , print "Access Granted. Door Unlocked." Otherwise, print "Access Denied.
Security Alerted.

```python
User_Code= int(input("Please Enter 4 digit Code"))
Check_Code= 1234

if User_Code == Check_Code:
        print("Access Granted")
else:
        print("Access Denied. Security Alerted")
```

2. The Weekend Grocery Trip A person is deciding whether to go to the supermarket.
Input: User enters how many eggs are left in the fridge (integer).
Scenario: If the number of eggs is less than 3, print "Time to go to the market!" Else, print "You have
enough for breakfast today.

```python
User_Input = int(input("Tell me how many number of EGGs are in Fridge"))
Egg_Threshold = 3
if User_Input < Egg_Threshold:
     print("Time to go to the market!")
else:
    print("You have enough for breakfast")
```

3. The Smart Home "Leaving for Work" Checklist When you leave your house, a smart assistant checks multiple
independent things. These are not mutually exclusive.
Input: Three separate inputs (Yes/No) for: "Is the AC on?", "Is the Coffee Maker on?", "Is the Front Door
unlocked?"

Scenario:
1. If AC is on Print "Turning off AC to save energy."
2. If Coffee Maker is on Print "Warning: Coffee Maker is still hot!"
3. If Door is unlocked Print "Locking the front door now." (Note: All three messages could potentially print.)

```python
ac = input("Is the AC on? (Yes/No): ").strip().lower()
coffe = input("Is the coffe machine on? (Yes/No): ").strip().lower()
door = input("Is the AC on? (Yes/No): ").strip().lower()

if ac and coffe and door!="yes":
    print(f"Please enter valid input {"Yes"} or {"No"} only")
if ac=="yes":
    print("Turning off AC to save energy")
if coffe=="yes":
    print("Warning: Coffee Maker is still hot!")
if door == "yes":
    print("Locking the front door now.")
```
